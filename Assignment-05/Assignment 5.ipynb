### Find out the definition of the above 9 variables and document the definitions in the Jupyter Notebook in a Markdown cell so that the readers of your notebook will know the meaning of these variables. 

UNITID- a unique identification number assigned to postsecondary institutions

INSTNM - The institution’s name.

CITY- CITY location belonging to institution.

STABBR - State belonging to institution.

ZIP - Zip code belonging to institution.

ADM_RATE - represents the admissions rate across all campuses, defined as the total number of admitted undergraduates across all branches divided by the total number of undergraduates who applied across all branches.

UGDS - includes the number of degree/certificate-seeking undergraduates enrolled in the fall.

TUITIONFEE_IN - the cost data include the tuition and required fees of the institution. This defines it as in state tuition only.

MN_EARN_WNE_P10- earnings are for the institutional aggregate of all federally aided students who enroll in an institution each year and who are employed but not enrolled.

### Find UMBC, UMCP, Loyola University and JHU. 

```python
umbc = data[data['INSTNM'] == 'University of Maryland-Baltimore County']
umbc    #Find UMBC
```

```python
umcp = data[data['INSTNM'] == 'University of Maryland-College Park']
umcp    #Find UMCP
```

```python
loyola = data[data['INSTNM'] == 'Loyola University Maryland']
loyola
```

```python
jhu = data[data['INSTNM'] == 'Johns Hopkins University']
jhu
```

### Calculate the min, max, and average in-state tution of the above four institutions. 

```python
four_university = umbc.append(umcp,ignore_index=True).append(loyola,ignore_index=True).append(jhu,ignore_index=True)
four_university
```

```python
four_university['TUITIONFEE_IN'].min()  #The min in-state tution of the above four institutions is $10,399.
```

```python
four_university['TUITIONFEE_IN'].max()  #The max in-state tution of the above four institutions is $52,170.
```

```python
four_university['TUITIONFEE_IN'].mean()  ##The average in-state tution of the above four institutions is $30,411.75.
```
