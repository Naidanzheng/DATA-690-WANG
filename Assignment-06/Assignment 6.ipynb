###  5. Write a function named cleanse_year() that takes a string as input and removes any asterisks in the string and return the cleansed string.

```python
def cleanse_year(raw_year):
  new_year=raw_year.replace('*','')           #replace asterisk with nothing
  return new_year
```

```python
cleanse_year("1989*")   #testing.
```

```python
cleanse_year("*20*10")
```

### 6.Cleanse the column "Census Year" using the function cleanse_year() and Pandas's apply() function. 

```python
data["Census Year"]=data["Census Year"].apply(cleanse_year)   #replace
data
```

### 7.Repeat this process for column "Total Population", "Census Cost", "Average Cost Per Person". 

```python
def cleanse_cost(raw_cost):
  new_cost=raw_cost.replace('$','').replace(',','').replace(' Billion','')  
  if raw_cost.find('4.5') != -1:                                            
    new_cost='4500000000'
  elif raw_cost.find('13') != -1:                                           
    new_cost='13000000000'
  return new_cost
```

```python
cleanse_cost('$4.5 Billion')        #testing
```

```python
cleanse_cost('$13 Billion')         #testing
```

```python
data["Census Cost"]=data["Census Cost"].apply(cleanse_cost)       
data  #replace
```

```python
def cleanse_avg(raw_avg):
  
  if raw_avg.find('cents') != -1:                    
    new_avg=float(raw_avg.replace('cents',''))/100   
  elif raw_avg.find('$') != -1:                      
    new_avg=raw_avg.replace('$','')                  #replace it 
  else: new_avg=raw_avg
  return new_avg
```

```python
cleanse_avg('1.13 cents')
```

```python
cleanse_avg('$42.11')
```

```python
data["Average Cost Per Person"]=data["Average Cost Per Person"].apply(cleanse_avg)  
data       #cleansed avg costs
```
